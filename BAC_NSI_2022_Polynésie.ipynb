```python
def racourcir(chaine):
  n=len(chaine)
  out=""
  f=1
  while f!=n-1:
    out=out+chaine[f]
    f+=1
  return out
```

```python
racourcir("frere")
```

```python
def regleA(chaine):
  n = len(chaine)
  if n >= 2:
    return chaine[0] == "a" and chaine[n-1] == "a" and regleA(racourcir(chaine))
  else:
    return chaine == "a"
```
